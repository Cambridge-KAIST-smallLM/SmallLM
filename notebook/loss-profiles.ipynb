```python
#MULTIPLE MODELS 
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df_123tied = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/123m_tied.csv")
df_162untied = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/162m_untied.csv")

# Plot the graph
plt.figure(figsize=(10, 6))

# Plot 123m Tied model
plt.plot(
    df_123tied['train/global_step'], 
    df_123tied['SmolLM-14m-Dolma-v0.1-Base-tied-123m-lr-6e4 - train/loss'], 
    label='123m Tied', 
    color='blue', 
    marker='o'
)

# Plot 162m Untied model
plt.plot(
    df_162untied['train/global_step'], 
    df_162untied['SmolLM-14m-Dolma-v0.1-Base-untied-162m-lr-6e4 - train/loss'], 
    label='162m Untied', 
    color='red', 
    marker='s'
)


# Labels and title
plt.xlabel("Global Step")
plt.ylabel("Training Loss")
plt.title("Effect of Weight Tying on Loss Profile")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
plt.savefig("tying.png")

```

```python
# MULTIPLE MODELS
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df_123tied = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/123m_tied.csv")
df_162untied = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/162m_untied.csv")
df_162untiedzloss = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/162m_tied_zloss.csv")
df_123tiedzloss = pd.read_csv("/Users/suchirsalhan/PHD/PHD/Notebook/123m_tied_zloss.csv")

# Plot the graph
plt.figure(figsize=(12, 7))

# Plot 162m Tied + Z loss
plt.plot(
    df_162untiedzloss['train/global_step'], 
    df_162untiedzloss['SmolLM-14m-Dolma-v0.1-Base-untied-162m-zloss-lr6e4 - train/loss'], 
    label='162m Untied + Z loss', 
    color='green', 
    marker='^'
)

# Plot 123m Tied + Z loss
plt.plot(
    df_123tiedzloss['train/global_step'], 
    df_123tiedzloss['SmolLM-14m-Dolma-v0.1-Base-tied-123m-zloss-lr6e4 - train/loss'], 
    label='123m Tied + Z loss', 
    color='purple', 
    marker='D'
)

# Labels and title
plt.xlabel("Global Step")
plt.ylabel("Training Loss")
plt.title("Loss Profiles of Llama Architecture with Z-Loss Auxiliary")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
plt.savefig("z-loss.png")
```
